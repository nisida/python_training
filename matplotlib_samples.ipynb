```python
import matplotlib.pyplot as plt
import numpy as np

data = {'a' : np.arange(50),
        'c' : np.random.randint(0, 50, 50),
        'd' : np.random.randn(50)}
#print("key a = ", data['a'])
#print("key c = ", data['c'])
#print("key d = ", data['d'])
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
#print("data['b'] = ", data['b'])
#print("data['d'] = ", data['d'])
plt.scatter('a', 'b', s='d', c='c', data=data)
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

name = ['group_a', 'group_b', 'group_c']
value = [1, 10, 100]

plt.figure(figsize=(9, 3))
#plt.subplot(131)
plt.bar(name, value)
plt.suptitle('Categorical Plotting')
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 0, 1
data = np.random.normal(mu, sigma, 1000)
print("data.shape = ", data.shape)

plt.hist(data, bins=40)
plt.title('Histogram')
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 5.0, 0.1)
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(t, np.exp(0 - t))
plt.title("subplot #1")
plt.subplot(2, 1, 2)
plt.plot(t, np.cos(2 * np.pi * t))
plt.title("subplot #2")
plt.xlabel('time (s)')
plt.show()
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig, axs = plt.subplots(2, 1)
axs[0].plot(x, y1)
axs[0].set_title('Sine Wave')
axs[1].plot(x, y2)
axs[1].set_title('Cosine Wave')
plt.tight_layout()
plt.show()
```
